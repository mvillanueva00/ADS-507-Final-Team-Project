```python
import pandas as pd
```

```python
import json

with open('/content/drug-ndc-0001-of-0001.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data['results'])
df.head()
```

```python
with open('/content/drug-shortages-0001-of-0001.json', 'r') as f:
    data = json.load(f)

df_shortages = pd.DataFrame(data['results'])
df_shortages.head()
```

```python
print("Columns in drug_products_ndc:")
print(df.columns.tolist())

print("\n" + "-"*50 + "\n")

print("Columns in drug_shortages:")
print(df_shortages.columns.tolist())
```

```python
# Size of Drug NDC dataset
print("Drug NDC Dataset Size")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
print(f"Memory usage (MB): {df.memory_usage(deep=True).sum() / 1024**2:.2f}")

print("\n" + "-"*50 + "\n")

# Size of Drug Shortages dataset
print("Drug Shortages Dataset Size")
print(f"Rows: {df_shortages.shape[0]}")
print(f"Columns: {df_shortages.shape[1]}")
print(f"Memory usage (MB): {df_shortages.memory_usage(deep=True).sum() / 1024**2:.2f}")
```
